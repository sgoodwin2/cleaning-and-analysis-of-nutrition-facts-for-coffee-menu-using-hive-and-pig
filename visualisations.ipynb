Import relevant libraries for data processing and visualisation

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt, figure
```

Import data from hive csv and give data headers

```python
df = pd.read_csv(r"/home/stang/CA4022-Starbucks-Nutrition-cleaning-analysis/data/archive/hive/000000_0", header=None)
df.columns = ['Catergory', 'Value']
```

Create bar chart to show comparison of fat content between chocolate and non-chocolate drinks

```python
sns.barplot(data=df, x = 'Catergory', y='Value')
plt.title('Average Fat Content in Grams')
plt.ylabel('Grams')
```
